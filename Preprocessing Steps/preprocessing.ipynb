```python
import numpy as np
import pandas as pd

from sklearn import preprocessing
```

#### Reading Cleaned data, as done in previous task

```python
travelData=pd.DataFrame(pd.read_excel('cleanedTravel.xlsx'))
```

#### Editing "Destination" Column to make it more relevant according to the required output

```python
travelData['Destination']=travelData['Destination']+','
travelData
```

```python
travelData['Destination'] = [travelData['Destination'].str.rsplit(',',2)[i][0] for i in range(136)]
```

```python
travelData
```

### Encoding

#### Label Encoding

Label encoding is a simple and effective way to convert categorical variables into numerical form.

Label Encoding is performed on the columns:
- Destination
- Traveler Gender
- Traveler Nationality
- Accommodation type
- Transportation type


```python
label_encoder = preprocessing.LabelEncoder()

travelData['Destination'] = label_encoder.fit_transform(travelData['Destination'])
travelData['Traveler_gender'] = label_encoder.fit_transform(travelData['Traveler_gender'])
travelData['Traveler_nationality'] = label_encoder.fit_transform(travelData['Traveler_nationality'])
travelData['Accommodation_type'] = label_encoder.fit_transform(travelData['Accommodation_type'])
travelData['Transportation_type'] = label_encoder.fit_transform(travelData['Transportation_type'])

travelData
```

#### One Hot Encoding

One Hot Encoding is a bit complex but is a more efficient way to convert categorical variables into numerical form in machine learning in which more space is used for the dummy array created for each features.
In this new features are added in terms of values of the old features.
